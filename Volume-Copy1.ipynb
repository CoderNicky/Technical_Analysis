```python
# Volume Analysis

import pandas as pd
import matplotlib.pyplot as plt
import os
from dotenv import load_dotenv
from alpha_vantage.timeseries import TimeSeries

load_dotenv()
alpha_vantage_api_key = os.getenv('ALPHA_VANTAGE_API_KEY')

# Set the API key as an environment variable
os.environ['ALPHAVANTAGE_API_KEY'] = 'VMNUT5346O7EZNEX'

# Create the TimeSeries object
ts = TimeSeries(output_format='pandas')
ts = TimeSeries(key=alpha_vantage_api_key, output_format='pandas')


tickers = ['TSLA'] #,'AAPL', 'GOOGL', 'MSFT', 'AMZN']
i = 0

for ticker in tickers:
    
    data, meta_data = ts.get_intraday(symbol=ticker, outputsize='full')   
    df = pd.DataFrame(data) #Create Panda Dataframe
    df = df.rename(columns={'1. open': 'Open', '4. close': 'Close','5. volume': 'Volume'}) #Rename Column
    print(df.head())  

# Plotting Price and Support and Resistance levels

    # Define a threshold for considering a volume significant
    volume_threshold = 1.5 * df['Volume'].mean()

    # Calculate support and resistance levels
    support_levels = df[df['Volume'] > volume_threshold]['Close']
    resistance_levels = df[df['Volume'] > volume_threshold]['Close']

    # Plotting price chart with support and resistance levels
    plt.figure(figsize=(12, 6))
    plt.plot(df['Close'], label='Close Price', linestyle='-', color='blue')
    plt.scatter(df.index[df['Volume'] > volume_threshold], df['Close'][df['Volume'] > volume_threshold], color='green', marker='o', label='Support/Resistance Level')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(str(tickers[i]) + ' Volume-Based Support and Resistance Levels')
    plt.legend()
    plt.grid(True)

# Display the calculated support and resistance levels
    print("Support Levels:")
    print(support_levels.unique())

    print("\nResistance Levels:")
    print(resistance_levels.unique())
    
    plt.figure(figsize=(12, 6))

    # Price Plot
    plt.subplot(2, 1, 1)
    plt.plot(df['Close'], label='Close Price', linestyle='-')
    plt.ylabel('USD Price')
    plt.title(str(tickers[i]) + ' Price and Volume Analysis')

# Plotting Price and Volume together

    # Volume Plot
    plt.subplot(2, 1, 2)
    plt.bar(df.index, df['Volume'], label='Volume', color='purple', alpha=0.1)
    plt.xlabel('Date')
    plt.ylabel('Volume')

    plt.tight_layout()
    plt.grid(True)
    
# (VPT) The Volume Price Trend detects the trend's strength by accumulating the
# values of volume multiplied by the percentage change in price.
# If the price is making higher highs while the VPT line is making lower highs,
# it indicates a reversal or weakness in the current trend.
# Consistently negative VPT indicates a continuation of the downward trend. 

    # Calculate VPT 
    df['Price_Change'] = df['Close'].pct_change()
    df['VPT'] = (df['Volume'] * df['Price_Change']).cumsum()

    # Plot VPT
    plt.figure(figsize=(12, 6))
    plt.plot(df['VPT'], label='VPT', color='blue')
    plt.xlabel('Date')
    plt.ylabel('VPT')
    plt.title(str(tickers[i]) + ' Volume Price Trend (VPT) Indicator')
    plt.legend()
    plt.grid(True)

# (OBV) On-Balance Volume adds or subtracts volume based on whether the closing price
# is higher or lower than the previous day.
# When the OBV line is rising, buying volume is dominating selling volume, 
# indicating bullishness and positive price momentum.
    
    # Calculate OBV
    df['OBV'] = (df['Volume'] * ((df['Close'] - df['Close'].shift(1)) > 0).astype(int)).cumsum()

    # Plot OBV
    plt.figure(figsize=(12, 6))
    plt.plot(df['OBV'], label='OBV')
    plt.xlabel('Date')
    plt.ylabel('OBV')
    plt.title(str(tickers[i]) + ' On-Balance Volume')
    plt.legend()
    plt.grid(True)
    
# (VWAP) Volume Weighted Average Price calculates the average price of an asset
# weighted by volume to assess whether an asset is trading above or below its average price

# Calculate VWAP
    df['Cumulative_Volume'] = df['Volume'].cumsum()
    df['Cumulative_Volume_Price'] = (df['Volume'] * df['Close']).cumsum()
    df['VWAP'] = df['Cumulative_Volume_Price'] / df['Cumulative_Volume']

# Plot VWAP
    plt.figure(figsize=(12, 6))
    plt.plot(df['Close'], label='Close Price', linestyle='-', color='blue')
    plt.plot(df['VWAP'], label='VWAP', linestyle='--', color='orange')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(str(tickers[i]) + ' Price vs. Volume Weighted Average Price (VWAP)')
    plt.legend()
    plt.grid(True)
    
# Price gaps occur when the opening price of a trading session is significantly different
# from the previous closing price
# High volume during a gap can indicate strong market interest

    # Calculate price gaps
    df['Price_Gap'] = df['Open'] - df['Close'].shift(1)

    # Define a threshold for considering a gap significant (e.g., 1% of the previous close)
    gap_threshold = 0.01 * df['Close'].shift(1)

    # Create a column to mark significant gaps
    df['Significant_Gap'] = abs(df['Price_Gap']) > gap_threshold

    # Plotting significant gaps
    plt.figure(figsize=(12, 6))
    plt.plot(df['Close'], label='Close Price', linestyle='-')
    plt.scatter(df.index[df['Significant_Gap']], df['Close'][df['Significant_Gap']], color='red', marker='o', label='Significant Gap')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(str(tickers[i]) + ' Gap Analysis')
    plt.legend()
    plt.grid(True)

    # Display the dates of significant gaps
    significant_gap_dates = df.index[df['Significant_Gap']]
    print("Dates of Significant Gaps:")
    print(significant_gap_dates)

# Volume Oscillators compare short-term and long-term volume averages
# to identify overbought or oversold conditions.

# Overbought Conditions:
# - When the oscillator crosses above a threshold (0 or a positive value),
# - Price is making higher highs, but the oscillator fails to make higher highs
#   and starts moving sideways or lower, it can indicate a potential reversal.
# - Extremely high values above its historical range) may signal overbought conditions. Traders often look for divergence or a subsequent downturn.
    
    # Calculate VPT Oscillator
    def calculate_vpt_oscillator(df, short_window, long_window):
        df['Price_Change'] = df['Close'].diff()
        df['VPT'] = (df['Volume'] * df['Price_Change']).cumsum()
    
        short_vpt = df['VPT'].rolling(window=short_window).mean()
        long_vpt = df['VPT'].rolling(window=long_window).mean()
    
        df['VPT_Oscillator'] = short_vpt - long_vpt
        return df

    short_window = 10  # Short-term window for VPT Oscillator
    long_window = 30  # Long-term window for VPT Oscillator

    df = calculate_vpt_oscillator(df, short_window, long_window)

# Calculate VWMA Oscillator
    def calculate_vwma_oscillator(df, short_window, long_window):
        df['VWMA_Short'] = (df['Volume'] * df['Close']).rolling(window=short_window).sum() / df['Volume'].rolling(window=short_window).sum()
        df['VWMA_Long'] = (df['Volume'] * df['Close']).rolling(window=long_window).sum() / df['Volume'].rolling(window=long_window).sum()
    
        df['VWMA_Oscillator'] = df['VWMA_Short'] - df['VWMA_Long']
        return df

    short_window = 10  # Short-term window for VWMA Oscillator
    long_window = 30  # Long-term window for VWMA Oscillator

    df = calculate_vwma_oscillator(df, short_window, long_window)

    # Plot VPT Oscillator
    plt.figure(figsize=(12, 6))
    plt.plot(df['VPT_Oscillator'], label='VPT Oscillator', color='blue')
    plt.axhline(y=0, color='red', linestyle='--', label='Zero Line')
    plt.xlabel('Date')
    plt.ylabel('VPT Oscillator')
    plt.title(str(tickers[i]) + ' Volume Price Trend (VPT) Oscillator')
    plt.legend()
    plt.grid(True)

    # Plot VWMA Oscillator
    plt.figure(figsize=(12, 6))
    plt.plot(df['VWMA_Oscillator'], label='VWMA Oscillator', color='green')
    plt.axhline(y=0, color='red', linestyle='--', label='Zero Line')
    plt.xlabel('Date')
    plt.ylabel('VWMA Oscillator')
    plt.title(str(tickers[i]) + ' Volume Weighted Moving Average (VWMA) Oscillator')
    plt.legend()
    plt.grid(True)

    plt.show()
    
    i = i + 1
```
